```python
def fill_jug1(state, cap1):
    x, y = state
    return (cap1, y)

def fill_jug2(state, cap2):
    x, y = state
    return (x, cap2)

def empty_jug1(state):
    x, y = state
    return (0, y)

def empty_jug2(state):
    x, y = state
    return (x, 0)

def pour_jug1_to_jug2(state, cap2):
    x, y = state
    transfer = min(x, cap2 - y)
    return (x - transfer, y + transfer)

def pour_jug2_to_jug1(state, cap1):
    x, y = state
    transfer = min(y, cap1 - x)
    return (x + transfer, y - transfer)

def dfs(cap1, cap2, target):

    stack = [(0, 0)]
    visited = set()

    while stack:

        state = stack.pop()

        if state in visited:
            continue

        visited.add(state)

        x, y = state
        print("Current State:", state)

        if x == target or y == target:
            print("Target Achieved")
            return

        s1 = fill_jug1(state, cap1)
        print("Fill Jug1 ->", s1)
        stack.append(s1)

        s2 = fill_jug2(state, cap2)
        print("Fill Jug2 ->", s2)
        stack.append(s2)

        s3 = empty_jug1(state)
        print("Empty Jug1 ->", s3)
        stack.append(s3)

        s4 = empty_jug2(state)
        print("Empty Jug2 ->", s4)
        stack.append(s4)

        s5 = pour_jug1_to_jug2(state, cap2)
        print("Pour Jug1 to Jug2 ->", s5)
        stack.append(s5)

        s6 = pour_jug2_to_jug1(state, cap1)
        print("Pour Jug2 to Jug1 ->", s6)
        stack.append(s6)

jug1 = int(input("Enter capacity of Jug1: "))
jug2 = int(input("Enter capacity of Jug2: "))
target = int(input("Enter target water amount: "))

dfs(jug1, jug2, target)
```
