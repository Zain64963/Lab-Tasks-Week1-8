```python
def solve_nqueens(N):
    solutions = []

    def dfs(row, cols, diag1, diag2, path):
        if row == N:
            board = []
            for c in path:
                row_str = ['.'] * N
                row_str[c] = 'Q'
                board.append(''.join(row_str))
            solutions.append(board)
            return
        for col in range(N):
            if not (cols & (1 << col) or diag1 & (1 << (row + col)) or diag2 & (1 << (row - col + N - 1))):
                dfs(row + 1,
                    cols | (1 << col),
                    diag1 | (1 << (row + col)),
                    diag2 | (1 << (row - col + N - 1)),
                    path + [col])

    dfs(0, 0, 0, 0, [])
    return solutions


N = int(input("N  size  (e.g., 4 ya 8): "))
sol = solve_nqueens(N)
print("Total solutions:", len(sol))


for i, board in enumerate(sol[:3], start=1):
    print(f"Solution {i}:")
    for row in board:
        print(row)
    print()
```
